3. Realice el pre-procesamiento que considere oportuno.

Para crear los embeddings, debemos tokenizar los títulos de las noticias, quitando los caracteres especiales, los espacios, y sustituyendo las letras que tienen tilde por esa misma letra sin la tilde. Además, eliminamos las "stopwords" en español. Transformamos las palabras para que estén todas las lletras en minúsculas y las tokenizamos. Así creamos el corpus con las frases para cada noticia de los cuales se generarán los embeddings. 

```python
wpt = nltk.WordPunctTokenizer()
stop_words = nltk.corpus.stopwords.words('spanish')

a,b = 'áéíóúü','aeiouu' ## 
trans = str.maketrans(a,b)

def normalize_document(doc):
    # se pasa a minúsculas, se eliminan caracteres especiales y espacios
    doc = doc.translate(trans)
    doc = re.sub(r'[^a-zA-Z\s]', '', doc, re.I|re.A)
    doc = doc.lower()
    doc = doc.strip()
    # tokenizado
    tokens = wpt.tokenize(doc)
    # filtro de stopwords: sólo se añaden los tokens que no figuran en la lista de stop-words
    filtered_tokens = [token for token in tokens if token not in stop_words]
    # Y se vuelve a crear el documento doc antes de devolverlo
    doc = ' '.join(filtered_tokens)
    return doc

normalize_corpus = np.vectorize(normalize_document)
```

```python
corpus = normalize_corpus(datos.Titular)
corpus
```

# Word Embeddings

## Aplicamos word2vec sobre el corpus

Como el corpus es pequeño, se estrenarán los embeddings sobre el propio corpus generado de las noticias. 

* Seleccionamos mayor feature size (20) para aumentar la variabilidad de pesos de cada palabra y mejorar la precisión de la segmentación. 
* Disminuimos también el window context a 3, para que el modelo no tenga en cuenta tantas palabras de contexto.
* Aumentamos el número mínimo de palabras que deben aparecer en el corpus para ser utilizadas en los embeddings, para eliminar nombres propios o palabras poco frecuentes. 
* Creamos el modelo aplicando el word2vec, con el corpus tokenizado y 100 iteraciones para mejorar el resultado.

```python
wpt = nltk.WordPunctTokenizer()
tokenized_corpus = [wpt.tokenize(document) for document in corpus]

# Ajustamos los valores de los parámetros del entrenamiento

feature_size = 20       # Dimensión de los vectores que constituyen los embeddings  
window_context = 6     # Tamaño de la "Ventana de Contexto", número de palabras "vecinas" que analizamos                                                                       
min_word_count = 2      # Si una palabra aparece menos veces en el corpus, no se utiliza para los embeddings                        
sample = 1e-3           # Las palabras con una frecuencia superior se disminuyen en la muestra

w2v_model = word2vec.Word2Vec(tokenized_corpus, size = feature_size, 
                              window = window_context, min_count = min_word_count,
                              sample = sample, iter = 100)
```

## Visualización de los embeddings

Los embeddings son vectores de n dimensiones, en nuestro caso 10. Queremos visualizarlos en un el plano en dos dimensiones, en un diagrama XY, así que utilizaremos TSNE, que es un visualizador de sklearn

```python
words = w2v_model.wv.index2word
wvs = w2v_model.wv[words]

tsne = TSNE(n_components = 2, random_state = 0, n_iter = 5000, perplexity = 2)
np.set_printoptions(suppress = True)
T = tsne.fit_transform(wvs)
labels = words

plt.figure(figsize = (12, 6))
plt.scatter(T[:, 0], T[:, 1], c = 'orange', edgecolors = 'r')
for label, x, y in zip(labels, T[:, 0], T[:, 1]):
    plt.annotate(label, xy = (x + 1, y + 1), xytext = (0, 0), textcoords = 'offset points')
```

## Embedding de un término

5. Construya embeddings a nivel de documento, sobre los embeddings de palabras anteriores.

Si solicitamos una palabra dentro del modeo de embeddings, podemos averiguar el peso que tiene esta palabra dentro del documento, es decir, la relevancia que puede tener esa palabra en el "diccionario" que se ha generado. Para adjudicar este peso se tienen en cuenta cosas como las veces que aparecen, la relevancia en el contexto en el que se encuentra, etc. Estos pesos se pueden utilizar para operar con ellos y son en los que se basa el siguiente embedding para realizar la segmentación, pues existe la opción de comparar la similaridad entre los pesos de todas las palabras.

```python
w2v_model.wv['nba']
```

Generamos las funciones que se utilizarán para obtener los embeddings de los términos y vectorizar las palabras. 

```python
## Build framework for getting document level embeddings

def average_word_vectors(words, model, vocabulary, num_features):
    
    feature_vector = np.zeros((num_features,),dtype="float64")
    nwords = 0.
    
    for word in words:
        if word in vocabulary: 
            nwords = nwords + 1.
            feature_vector = np.add(feature_vector, model[word])
    
    if nwords:
        feature_vector = np.divide(feature_vector, nwords)
        
    return feature_vector
```

```python
def averaged_word_vectorizer(corpus, model, num_features):
    vocabulary = set(model.wv.index2word)
    features = [average_word_vectors(tokenized_sentence, model, vocabulary, num_features)
                    for tokenized_sentence in corpus]
    return np.array(features)
```

Generamos la matriz de características donde comparamos los pesos de las palabras con el corpus tokenizado.

```python
w2v_feature_array = averaged_word_vectorizer(corpus=tokenized_corpus, model=w2v_model,
                                             num_features=feature_size)
pd.DataFrame(w2v_feature_array)
```

### Affinity Propagation

```python
ap = AffinityPropagation()
ap.fit(w2v_feature_array)
cluster_labels = ap.labels_
cluster_labels = pd.DataFrame(cluster_labels, columns=['ClusterLabel'])
pd.concat([datos, cluster_labels], axis=1)
```

```python
pca = PCA(n_components=2, random_state=0)
pcs = pca.fit_transform(w2v_feature_array)
labels = ap.labels_
categories = list(datos['Etiquetas'])
plt.figure(figsize=(8, 6))

for i in range(len(labels)):
    label = labels[i]
    color = 'orange' if label == 0 else 'blue' if label == 1 else 'green' if label == 2 else 'pink' if label == 3 else 'purple' if label == 4 else 'yellow'  
    annotation_label = categories[i]
    x, y = pcs[i]
    plt.scatter(x, y, c=color, edgecolors='k')
    plt.annotate(annotation_label, xy=(x+1e-4, y+1e-3), xytext=(0, 0), textcoords='offset points')
```

### K-Means

Seleccionamos una generación de tan solo 3 clústeres, a pesar de que en el modelo de segmentación anterior salieron como los más óptimos 6 clústeres, porque queremos que agrupe los títulos de la noticia de forma que concuerde con la etiqueta adjudicada de cada uno.

```python
km = KMeans(n_clusters = 3, random_state=0)
km.fit_transform(w2v_feature_array)
cluster_labels = km.labels_
cluster_labels = pd.DataFrame(cluster_labels, columns=['ClusterLabel'])
pd.concat([datos, cluster_labels], axis=1)
```

## Comentarios Word2Vec

Como comentario de esta parte, vemos como el primer modelo de segmentación tiene dos desventajas: 
1. No nos permite elegir el número de clústeres que deseamos generar y 
2. crea demasiados que, a la hora de graficarlos, vemos como no coinciden con las etiquetas en cuestión. 

En algún clúster, como el azul, el rosa o el naranja, si que identifica mayoritariamente la etiqueta internacional (con allgún nacional) pero la etiqueta de deportes, por ejemplo, está mezclada en casi todos los grupos. No se observa que los documentos de ninguna estiqueta estén más cercanos a los demás de su misma etiqueta. 

Por otro lado, con el método de K-means podemos mencionar un pro y un contra: 
* Positivo: podemos seleccionar el número de clústers que deseamos generar. 
* Contra: no aciertan las etiquetas. 

Como en el caso anterior, las tres etiquetas aparecen mezcladas en el gráfico y no se agrupan entre ellas en función con su etiqueta. Además, los tres clústeres mezclan títulos de las tres etiquetas, con lo cual, el objetivo no se ha alcanzado. 

Cabe destacar también que, en Word2Vec, el número de iteraciones en la creación de embeddings influye en el número de clústeres que salen, pero no en la precisión para segmentar. Sin embarg, en GloVe no existe ese problema.

En un corpus tan breve no aporta ventajas claras utilizar embeddings frente a la alternativa de obtener una matriz Tf-Idf y luego una matriz de similitud. Como se ha utilizado un modelo sin entrenar y el entrenamiento se ha realizado sobre nuestro corpus, hemos podido utilizar Word2Vec en español sin problemas. Existe la opción de cargar un modelo ya entrenado, y en ese caso sí que tendríamos que tener en cuenta el idioma, y descargar el modelo adecuado.

## Comentarios GloVe

En Glove ambos modelos tienen los mismos problemas y restricciones que en Word2Vec, como vemos a continuación:

__Primer modelo:__
* No es posible fijar el número de clústers 
* No segmenta bien (las etiquetas están mezcladas entre ellas)

__Segundo caso:__
* Permite elegir el número de clústeres que queremos generar
* Tampoco segmenta bien (se esperaba que los clústeres generados coincidiesen con su etiqueta, pero no es el caso).

Sin embargo, cabe mencionar que la precisión a la hora de segmentar es un poco mejor en GloVe que en Word2Vec. Vemos como, por ejemplo, los títulos con etiquetas de Internacional o Nacional están cerca unos de otros. Deportes no se identifica bien en ninguno de los dos métodos (probablemente porque haya noticias de deporte tanto nacionales como internacionales, y siempre tenderá a agruparlo con uno de ellos). 

Como conclusión, es probable que la segmentación no es muy precisa en ninguno de los casos porque el corpus con el que se está trabajando es muy pequeñño. Si se aumentase el tamaño del corpus, es probable que mejorasen las segmentaciones. Además, segmentan mejor ambos modelos aplicando el método de GloVe.
