## CON NLTK

### Texto en español

1. Obtenga dos ficheros en formato txt, uno de ellos en ingles y otro en __español__, que contengan textos con entidades a  extraer. Busque textos con un conjunto amplio de entidades, no se limite a dos o tres de cada tipo. Empezamos la práctica con el texto seleccionado en español (una noticia sobre el coronavirus del periodico EL MUNDO).

```python
# Importamos las librerías necesarias e introducimos un texto

import nltk, re, pprint
nltk.download('maxent_ne_chunker')
nltk.download('words')
import pandas  as pd

texto = open('texto_esp.txt', encoding = 'UTF-8')
texto = texto.read()
```

2. Lea estos textos desde un programa Python, y realice las labores de pre-procesamiento habituales: división en frases,  división en palabras y conversión a las formas normales. Utilice los frameworks que prefiera. Elegimos primero hacer NLTK.

```python
# Dividimos en frases

frases = nltk.sent_tokenize(texto, language = 'spanish')
print(frases)
```

```python
# Tokenizamos

frases_token = [nltk.word_tokenize(frase) for frase in frases]
print(frases_token)
```

3. Realice un análisis morfológico de los términos incluidos. 

```python
# Análisis morfológico

frases_tagged = [nltk.pos_tag(frase) for frase in frases_token]
print(frases_tagged)
```

4. Por último, extraiga entidades de los textos anteriores

```python
# Extracción de entidades

for fras in frases_tagged:
    print(nltk.ne_chunk(fras))
```

## CON SPACY

### En español

1. Obtenga dos ficheros en formato txt, uno de ellos en ingles y otro en __español__, que contengan textos con entidades a  extraer. Busque textos con un conjunto amplio de entidades, no se limite a dos o tres de cada tipo. Continuamos la práctica con el texto seleccionado en español (una noticia sobre el coronavirus del periodico EL MUNDO).

```python
# Importamos las librerías que corresponda
# Hay que añadir los modelos que se utilicen. Previamente cargados con python -m spacy download es_core_news_md

import spacy
from spacy import displacy

nlp_esp = spacy.load('es_core_news_md')
# nlp_eng = spacy.load('en_core_web_sm')
```

2. Lea estos textos desde un programa Python, y realice las labores de pre-procesamiento habituales: división en frases,  división en palabras y conversión a las formas normales. Utilice los frameworks que prefiera. Elegimos ahora aplicar spaCy.

```python
# Leemos un texto

docum = open('texto_esp.txt', encoding = 'UTF-8')
docum = docum.read()
```

```python
# Ejecutamos la pipeline de análisis

docum_analizado = nlp_esp(docum)
```

```python
# Imprimimos frases

for sent in docum_analizado.sents:
    print(sent.text)
```

3. Realice un análisis morfológico de los términos incluidos. 

```python
# Imprimimos tokens, formas canónicas y entidades reconocidas


# POS tagging with Spacy 
spacy_pos_tagged = [(palabra.text, palabra.lemma_, palabra.pos_) for palabra in docum_analizado]
pd.DataFrame(spacy_pos_tagged, columns=['Palabra', 'F. Canónica', 'POS']).T
```

4. Por último, extraiga entidades de los textos anteriores

```python
# Imprimimos entidades

ents_tagged = [(ent.text, ent.label_) for ent in docum_analizado.ents]
pd.DataFrame(ents_tagged, columns=['Entidad', 'Tipo']).T
```

```python
# Entidades en el texto

for sent in docum_analizado.sents:
    displacy.render(sent, style='ent', jupyter=True, options={'distance': 110,'arrow_stroke': 2,'arrow_width': 8})
```

```python
# Diagrama de Dependencias

for sent in docum_analizado.sents:
    displacy.render(sent, jupyter=True, options={'distance': 110,'arrow_stroke': 2,'arrow_width': 8})
```

5. Una vez realizados todos los análisis, liste en pantalla las anotacioes obtenidas: frases, palabras, categoría  gramatical del análisis morfológico, localizaciones, organizaciones y personas

```python
tabla_entidades_esp = pd.DataFrame(ents_tagged, columns=['Entidad', 'Tipo'])
tabla_entidades_esp.groupby(['Tipo','Entidad']).count()
```

En primer lugar, cabe explicar la diferencia entre los resultados obtenidos del texto en inglés y del texto en español. Tanto utilizando NLTK como spaCy, el programa identifica mucho mejor las entidades en el texto en inglés que en el español. Con ambos métodos, a la hora de identificar las entidades en español, comete errores como confundir 'Sanidad' (de Ministerio de Sanidad) con una persona o considerar el Gobierno o algún Ministerio con un lugar. A la hora de utilizar un método u otro, los dos consiguen exitosamente encontrar una serie de entidades, para un idioma mejor que para otro, como explicado anteriormente. Sin embargo, el método de spaCy, aparte de devolver resultados más precisos, nos muestra la información de forma mucho mas entendible y más fácil de manipular. Personalmente, encuentro mejor el método de spaCy para este tipo de análisis, no solo muestra el análisis morfológico para cada frase, sino que genera un nuevo dataframe con el que trabajar y agrupar las entidades resulta mucho más cómodo.
